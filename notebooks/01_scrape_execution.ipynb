### Ensure everything is in order 
Check if src/ is importable when running from the notebooks. Seeing the root dir of the project will confirm that we are ready to proceed


```python
import sys
from pathlib import Path

ROOT = Path.cwd()
if ROOT.name == 'notebooks':
    ROOT = ROOT.parent
print(ROOT)
sys.path.append(str(ROOT))

from run_pilot_scrape import main as run_scrape
```

### Scrape Setup & Run

```python
pages = 100
delay = 0
log_every = 10
output_prefix = ROOT / 'data/raw/sales/raw_room'

run_scrape([
    f'--pages={pages}',
    f'--delay={delay}',
    f'--log-every={log_every}',
    f'--output-prefix={output_prefix}'
])
```
